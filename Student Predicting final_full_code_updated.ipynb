```python
# ============================================================
# TRAIN-TEST SPLIT, LABEL ENCODING, STANDARD SCALING AND MODELS
# Models used:
# 1. Artificial Neural Network
# 2. Support Vector Machine
# 3. Decision Tree
# ============================================================

# Machine learning imports
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler, label_binarize
from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix,
    ConfusionMatrixDisplay,
    roc_auc_score,
    roc_curve,
    auc
)

# Models
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier

# Extra import for plotting
import matplotlib.pyplot as plt
import numpy as np
```

```python
# ============================================================
# CREATE FEATURE MATRIX X AND TARGET VECTOR y
# ============================================================

# The original Target column is text-based and should not be used directly for modelling.
# Target_encoded was already created during EDA:
# Dropout = 0, Enrolled = 1, Graduate = 2

X = df.drop(columns=["Target", "Target_encoded"])
y = df["Target_encoded"].astype(int)

# Display feature and target shapes
print("Feature matrix shape:", X.shape)
print("Target vector shape:", y.shape)
```

```python
# ============================================================
# LABEL ENCODING FOR CATEGORICAL AND BINARY COLUMNS
# ============================================================

# Make a copy of X so the original dataframe remains unchanged
X_encoded = X.copy()

# Combine categorical and binary columns for label encoding
label_encode_cols = categorical_cols + binary_cols

# Dictionary to store fitted LabelEncoders for future use
label_encoders = {}

# Apply LabelEncoder to each categorical/binary column
for col in label_encode_cols:
    le = LabelEncoder()
    X_encoded[col] = le.fit_transform(X_encoded[col].astype(str))
    label_encoders[col] = le

# Confirm encoded dataset
X_encoded.head()
```

```python
# ============================================================
# TRAIN-TEST SPLIT
# ============================================================

# Stratify keeps the same class proportion in train and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
# ============================================================
# STANDARD SCALING
# ============================================================

# StandardScaler is fitted only on the training data to avoid data leakage.
# Scaled data will be used for ANN and SVM.
# Decision Tree does not require scaling, so it will use unscaled encoded data.

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Scaling completed successfully.")
```

```python
# ============================================================
# MODEL 1: ARTIFICIAL NEURAL NETWORK
# ============================================================

ann_model = MLPClassifier(
    hidden_layer_sizes=(128, 64, 32),
    activation="relu",
    solver="adam",
    alpha=0.0001,
    learning_rate="adaptive",
    learning_rate_init=0.001,
    batch_size=64,
    max_iter=1000,
    early_stopping=True,
    validation_fraction=0.15,
    n_iter_no_change=20,
    random_state=42
)

# Train ANN model
ann_model.fit(X_train_scaled, y_train)

# Predictions
ann_y_pred = ann_model.predict(X_test_scaled)
ann_y_proba = ann_model.predict_proba(X_test_scaled)

# Evaluation
print("ANN Accuracy:", accuracy_score(y_test, ann_y_pred))

print("\nANN Classification Report:")
print(classification_report(
    y_test,
    ann_y_pred,
    target_names=["Dropout", "Enrolled", "Graduate"]
))
```

```python
# ============================================================
# ANN LOSS CURVE ACROSS EPOCHS / ITERATIONS
# ============================================================

plt.figure(figsize=(8, 5))

# Training loss curve from ANN model
plt.plot(
    ann_model.loss_curve_,
    label="Training Loss",
    linewidth=2
)

# Validation curve is available only if early_stopping=True was used in ANN
if hasattr(ann_model, "validation_scores_"):
    validation_error = 1 - np.array(ann_model.validation_scores_)

    plt.plot(
        validation_error,
        label="Validation Error",
        linewidth=2
    )

plt.title("ANN Training Loss Across Epochs")
plt.xlabel("Epochs / Iterations")
plt.ylabel("Loss / Error")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# ============================================================
# MODEL 2: SUPPORT VECTOR MACHINE
# ============================================================

svm_model = SVC(
    kernel="rbf",
    C=1.0,
    gamma="scale",
    probability=True,
    random_state=42
)

# Train SVM model
svm_model.fit(X_train_scaled, y_train)

# Predictions
svm_y_pred = svm_model.predict(X_test_scaled)
svm_y_proba = svm_model.predict_proba(X_test_scaled)

# Evaluation
print("SVM Accuracy:", accuracy_score(y_test, svm_y_pred))
print("\nSVM Classification Report:")
print(classification_report(y_test, svm_y_pred, target_names=["Dropout", "Enrolled", "Graduate"]))
```

```python
# ============================================================
# MODEL 3: RANDOM FOREST
# ============================================================

from sklearn.ensemble import RandomForestClassifier

random_forest_model = RandomForestClassifier(
    n_estimators=200,
    criterion="gini",
    max_depth=None,
    min_samples_split=2,
    min_samples_leaf=1,
    random_state=42,
    n_jobs=-1
)

# Train Random Forest model
random_forest_model.fit(X_train, y_train)

# Predictions
random_forest_y_pred = random_forest_model.predict(X_test)
random_forest_y_proba = random_forest_model.predict_proba(X_test)

# Evaluation
print("Random Forest Accuracy:", accuracy_score(y_test, random_forest_y_pred))
print("\nRandom Forest Classification Report:")
print(classification_report(
    y_test,
    random_forest_y_pred,
    target_names=["Dropout", "Enrolled", "Graduate"]
))
```

```python
# ============================================================
# CONFUSION MATRIX FUNCTION
# ============================================================

def plot_confusion_matrix(y_true, y_pred, model_name):
    """
    This function plots the confusion matrix for a trained model.
    """
    cm = confusion_matrix(y_true, y_pred)

    fig, ax = plt.subplots(figsize=(6, 5))

    disp = ConfusionMatrixDisplay(
        confusion_matrix=cm,
        display_labels=["Dropout", "Enrolled", "Graduate"]
    )

    disp.plot(cmap="Blues", values_format="d", ax=ax)
    plt.title(f"Confusion Matrix - {model_name}")
    plt.show()


# ============================================================
# CONFUSION MATRICES FOR ALL MODELS
# ============================================================

plot_confusion_matrix(
    y_test,
    ann_y_pred,
    "Artificial Neural Network"
)

plot_confusion_matrix(
    y_test,
    svm_y_pred,
    "Support Vector Machine"
)

plot_confusion_matrix(
    y_test,
    random_forest_y_pred,
    "Random Forest"
)
```

```python
# ============================================================
# MULTI-CLASS ROC-AUC FUNCTION
# ============================================================

def plot_multiclass_roc_auc(y_true, y_proba, model_name):
    """
    This function plots ROC curves and calculates ROC-AUC scores
    for multi-class classification using One-vs-Rest strategy.
    """

    # Define class labels
    classes = [0, 1, 2]
    class_names = ["Dropout", "Enrolled", "Graduate"]

    # Binarize true labels for multi-class ROC calculation
    y_true_binarized = label_binarize(y_true, classes=classes)

    # Calculate overall macro ROC-AUC score
    macro_auc = roc_auc_score(
        y_true_binarized,
        y_proba,
        average="macro",
        multi_class="ovr"
    )

    print(f"{model_name} Macro ROC-AUC Score:", round(macro_auc, 4))

    # Plot ROC curve for each class
    plt.figure(figsize=(8, 6))

    for i, class_name in enumerate(class_names):
        fpr, tpr, _ = roc_curve(y_true_binarized[:, i], y_proba[:, i])
        roc_auc = auc(fpr, tpr)

        plt.plot(
            fpr,
            tpr,
            label=f"{class_name} AUC = {roc_auc:.4f}"
        )

    # Random classifier reference line
    plt.plot([0, 1], [0, 1], linestyle="--", label="Random Classifier")

    plt.title(f"ROC-AUC Curve - {model_name}")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.legend()
    plt.grid(True)
    plt.show()


# ============================================================
# ROC-AUC CURVES FOR ALL MODELS
# ============================================================

plot_multiclass_roc_auc(
    y_test,
    ann_y_proba,
    "Artificial Neural Network"
)

plot_multiclass_roc_auc(
    y_test,
    svm_y_proba,
    "Support Vector Machine"
)

plot_multiclass_roc_auc(
    y_test,
    random_forest_y_proba,
    "Random Forest"
)
```

```python
# ============================================================
#  HYPERPARAMETER TUNING USING RANDOMIZED SEARCH CV
# ANN, SVM, DECISION TREE
# ============================================================

from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import accuracy_score, classification_report, f1_score, roc_auc_score
from sklearn.preprocessing import label_binarize
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
```

```python
# ============================================================
# TUNING MODEL 1: ARTIFICIAL NEURAL NETWORK
# ============================================================

ann_tuning_model = MLPClassifier(
    max_iter=500,
    early_stopping=True,
    validation_fraction=0.15,
    n_iter_no_change=15,
    random_state=42
)

ann_param_dist = {
    "hidden_layer_sizes": [
        (64, 32),
        (128, 64),
        (128, 64, 32),
        (256, 128)
    ],
    "activation": ["relu", "tanh"],
    "solver": ["adam"],
    "alpha": [0.0001, 0.001, 0.01],
    "learning_rate": ["constant", "adaptive"],
    "learning_rate_init": [0.001, 0.0005],
    "batch_size": [64, 128]
}

ann_random = RandomizedSearchCV(
    estimator=ann_tuning_model,
    param_distributions=ann_param_dist,
    n_iter=8,
    scoring="f1_weighted",
    cv=2,
    n_jobs=-1,
    verbose=1,
    random_state=42
)

ann_random.fit(X_train_scaled, y_train)

best_ann_model = ann_random.best_estimator_

print("Best ANN Parameters:")
print(ann_random.best_params_)

ann_tuned_y_pred = best_ann_model.predict(X_test_scaled)
ann_tuned_y_proba = best_ann_model.predict_proba(X_test_scaled)

print("\nTuned ANN Accuracy:", accuracy_score(y_test, ann_tuned_y_pred))

print("\nTuned ANN Classification Report:")
print(classification_report(
    y_test,
    ann_tuned_y_pred,
    target_names=["Dropout", "Enrolled", "Graduate"]
))
```

```python
# ============================================================
# TUNING MODEL 2: SUPPORT VECTOR MACHINE
# ============================================================

svm_tuning_model = SVC(
    probability=False,
    random_state=42
)

svm_param_dist = {
    "C": [0.1, 1, 10, 50],
    "kernel": ["linear", "rbf"],
    "gamma": ["scale", "auto", 0.01, 0.1]
}

svm_random = RandomizedSearchCV(
    estimator=svm_tuning_model,
    param_distributions=svm_param_dist,
    n_iter=8,
    scoring="f1_weighted",
    cv=2,
    n_jobs=-1,
    verbose=1,
    random_state=42
)

svm_random.fit(X_train_scaled, y_train)

print("Best SVM Parameters:")
print(svm_random.best_params_)

# Refit final SVM with probability=True for ROC-AUC
best_svm_model = SVC(
    **svm_random.best_params_,
    probability=True,
    random_state=42
)

best_svm_model.fit(X_train_scaled, y_train)

svm_tuned_y_pred = best_svm_model.predict(X_test_scaled)
svm_tuned_y_proba = best_svm_model.predict_proba(X_test_scaled)

print("\nTuned SVM Accuracy:", accuracy_score(y_test, svm_tuned_y_pred))

print("\nTuned SVM Classification Report:")
print(classification_report(
    y_test,
    svm_tuned_y_pred,
    target_names=["Dropout", "Enrolled", "Graduate"]
))
```

```python
# ============================================================
# TUNING MODEL 3: RANDOM FOREST
# ============================================================

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import accuracy_score, classification_report

rf_tuning_model = RandomForestClassifier(
    random_state=42,
    n_jobs=-1
)

rf_param_dist = {
    "n_estimators": [100, 150, 200, 300],
    "criterion": ["gini", "entropy"],
    "max_depth": [5, 10, 15, 20, None],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4],
    "max_features": ["sqrt", "log2", None],
    "bootstrap": [True, False]
}

rf_random = RandomizedSearchCV(
    estimator=rf_tuning_model,
    param_distributions=rf_param_dist,
    n_iter=10,
    scoring="f1_weighted",
    cv=2,
    n_jobs=-1,
    verbose=1,
    random_state=42
)

# Train Random Forest tuning model
rf_random.fit(X_train, y_train)

# Best tuned Random Forest model
best_rf_model = rf_random.best_estimator_

print("Best Random Forest Parameters:")
print(rf_random.best_params_)

# Predictions
rf_tuned_y_pred = best_rf_model.predict(X_test)
rf_tuned_y_proba = best_rf_model.predict_proba(X_test)

# Evaluation
print("\nTuned Random Forest Accuracy:", accuracy_score(y_test, rf_tuned_y_pred))

print("\nTuned Random Forest Classification Report:")
print(classification_report(
    y_test,
    rf_tuned_y_pred,
    target_names=["Dropout", "Enrolled", "Graduate"]
))
```

```python
# ============================================================
# COMPARISON OF TUNED MODELS
# ============================================================

y_test_bin = label_binarize(y_test, classes=[0, 1, 2])

tuned_results = pd.DataFrame({
    "Model": [
        "Tuned ANN",
        "Tuned SVM",
        "Tuned Random Forest"
    ],
    "Accuracy": [
        accuracy_score(y_test, ann_tuned_y_pred),
        accuracy_score(y_test, svm_tuned_y_pred),
        accuracy_score(y_test, rf_tuned_y_pred)
    ],
    "Weighted F1 Score": [
        f1_score(y_test, ann_tuned_y_pred, average="weighted"),
        f1_score(y_test, svm_tuned_y_pred, average="weighted"),
        f1_score(y_test, rf_tuned_y_pred, average="weighted")
    ],
    "Macro ROC-AUC": [
        roc_auc_score(
            y_test_bin,
            ann_tuned_y_proba,
            average="macro",
            multi_class="ovr"
        ),
        roc_auc_score(
            y_test_bin,
            svm_tuned_y_proba,
            average="macro",
            multi_class="ovr"
        ),
        roc_auc_score(
            y_test_bin,
            rf_tuned_y_proba,
            average="macro",
            multi_class="ovr"
        )
    ]
})

tuned_results.round(4)
```

```python
# ============================================================
# TUNED MODEL PERFORMANCE VISUALIZATION
# ============================================================

plt.figure(figsize=(10, 5))

tuned_results_melted = tuned_results.melt(
    id_vars="Model",
    value_vars=["Accuracy", "Weighted F1 Score", "Macro ROC-AUC"],
    var_name="Metric",
    value_name="Score"
)

sns.barplot(
    data=tuned_results_melted,
    x="Model",
    y="Score",
    hue="Metric"
)

plt.title("Comparison of Tuned ANN, SVM and Random Forest")
plt.xlabel("Model")
plt.ylabel("Score")
plt.ylim(0, 1)
plt.xticks(rotation=20)
plt.legend(title="Metric")
plt.tight_layout()
plt.show()
```

```python
# ============================================================
# CONFUSION MATRICES FOR TUNED MODELS
# Tuned ANN, Tuned SVM, Tuned Random Forest
# ============================================================

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

class_names = ["Dropout", "Enrolled", "Graduate"]

tuned_predictions = {
    "Tuned ANN": ann_tuned_y_pred,
    "Tuned SVM": svm_tuned_y_pred,
    "Tuned Random Forest": rf_tuned_y_pred
}

for model_name, y_pred in tuned_predictions.items():
    cm = confusion_matrix(y_test, y_pred)

    plt.figure(figsize=(6, 5))
    disp = ConfusionMatrixDisplay(
        confusion_matrix=cm,
        display_labels=class_names
    )

    disp.plot(cmap="Blues", values_format="d")
    plt.title(f"Confusion Matrix - {model_name}")
    plt.xlabel("Predicted Class")
    plt.ylabel("Actual Class")
    plt.tight_layout()
    plt.show()
```

```python
# ============================================================
# ROC-AUC CURVES FOR TUNED MODELS
# Multi-class ROC-AUC using One-vs-Rest strategy
# ============================================================

from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc, roc_auc_score
import matplotlib.pyplot as plt
import numpy as np

classes = [0, 1, 2]
class_names = ["Dropout", "Enrolled", "Graduate"]

# Convert y_test into binary format for multi-class ROC
y_test_bin = label_binarize(y_test, classes=classes)

tuned_probabilities = {
    "Tuned ANN": ann_tuned_y_proba,
    "Tuned SVM": svm_tuned_y_proba,
    "Tuned Random Forest": rf_tuned_y_proba
}

for model_name, y_proba in tuned_probabilities.items():

    macro_auc = roc_auc_score(
        y_test_bin,
        y_proba,
        average="macro",
        multi_class="ovr"
    )

    print(f"{model_name} Macro ROC-AUC Score: {macro_auc:.4f}")

    plt.figure(figsize=(8, 6))

    for i, class_name in enumerate(class_names):
        fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_proba[:, i])
        class_auc = auc(fpr, tpr)

        plt.plot(
            fpr,
            tpr,
            label=f"{class_name} AUC = {class_auc:.4f}"
        )

    plt.plot(
        [0, 1],
        [0, 1],
        linestyle="--",
        label="Random Classifier"
    )

    plt.title(f"ROC-AUC Curve - {model_name}")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
