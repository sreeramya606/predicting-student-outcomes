```python
# Convert Target column to categorical type
df["Target"] = df["Target"].astype("category")

# Create numeric target column for correlation analysis later
# This is not model encoding yet, only useful for EDA
target_mapping = {
    "Dropout": 0,
    "Enrolled": 1,
    "Graduate": 2
}

df["Target_encoded"] = df["Target"].map(target_mapping)

df[["Target", "Target_encoded"]].head()
```

```python
# Convert categorical coded columns to category type
for col in categorical_cols:
    df[col] = df[col].astype("category")

# Convert binary columns to category type
for col in binary_cols:
    df[col] = df[col].astype("category")

# Keep numerical columns as numeric
for col in numerical_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# Check updated data types
df.dtypes
```

```python
# ============================================================
# TRAIN-TEST SPLIT, LABEL ENCODING, STANDARD SCALING AND MODELS
# Models used:
# 1. Artificial Neural Network
# 2. Support Vector Machine
# 3. Decision Tree
# ============================================================

# Machine learning imports
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler, label_binarize
from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix,
    ConfusionMatrixDisplay,
    roc_auc_score,
    roc_curve,
    auc
)

# Models
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier

# Extra import for plotting
import matplotlib.pyplot as plt
import numpy as np
```

```python
# ============================================================
# CREATE FEATURE MATRIX X AND TARGET VECTOR y
# ============================================================

# The original Target column is text-based and should not be used directly for modelling.
# Target_encoded was already created during EDA:
# Dropout = 0, Enrolled = 1, Graduate = 2

X = df.drop(columns=["Target", "Target_encoded"])
y = df["Target_encoded"].astype(int)

# Display feature and target shapes
print("Feature matrix shape:", X.shape)
print("Target vector shape:", y.shape)
```

```python
# ============================================================
# LABEL ENCODING FOR CATEGORICAL AND BINARY COLUMNS
# ============================================================

# Make a copy of X so the original dataframe remains unchanged
X_encoded = X.copy()

# Combine categorical and binary columns for label encoding
label_encode_cols = categorical_cols + binary_cols

# Dictionary to store fitted LabelEncoders for future use
label_encoders = {}

# Apply LabelEncoder to each categorical/binary column
for col in label_encode_cols:
    le = LabelEncoder()
    X_encoded[col] = le.fit_transform(X_encoded[col].astype(str))
    label_encoders[col] = le

# Confirm encoded dataset
X_encoded.head()
```

```python
# ============================================================
# TRAIN-TEST SPLIT
# ============================================================

# Stratify keeps the same class proportion in train and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
# ============================================================
# STANDARD SCALING
# ============================================================

# StandardScaler is fitted only on the training data to avoid data leakage.
# Scaled data will be used for ANN and SVM.
# Decision Tree does not require scaling, so it will use unscaled encoded data.

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Scaling completed successfully.")
```

```python
# ============================================================
# MODEL 1: ARTIFICIAL NEURAL NETWORK
# ============================================================

ann_model = MLPClassifier(
    hidden_layer_sizes=(64, 32),
    activation="relu",
    solver="adam",
    max_iter=500,
    random_state=42
)

# Train ANN model
ann_model.fit(X_train_scaled, y_train)

# Predictions
ann_y_pred = ann_model.predict(X_test_scaled)
ann_y_proba = ann_model.predict_proba(X_test_scaled)

# Evaluation
print("ANN Accuracy:", accuracy_score(y_test, ann_y_pred))
print("\nANN Classification Report:")
print(classification_report(y_test, ann_y_pred, target_names=["Dropout", "Enrolled", "Graduate"]))
```

```python
# ============================================================
# MODEL 2: SUPPORT VECTOR MACHINE
# ============================================================

svm_model = SVC(
    kernel="rbf",
    C=1.0,
    gamma="scale",
    probability=True,
    random_state=42
)

# Train SVM model
svm_model.fit(X_train_scaled, y_train)

# Predictions
svm_y_pred = svm_model.predict(X_test_scaled)
svm_y_proba = svm_model.predict_proba(X_test_scaled)

# Evaluation
print("SVM Accuracy:", accuracy_score(y_test, svm_y_pred))
print("\nSVM Classification Report:")
print(classification_report(y_test, svm_y_pred, target_names=["Dropout", "Enrolled", "Graduate"]))
```

```python
# ============================================================
# MODEL 3: DECISION TREE
# ============================================================

decision_tree_model = DecisionTreeClassifier(
    criterion="gini",
    max_depth=None,
    random_state=42
)

# Train Decision Tree model
decision_tree_model.fit(X_train, y_train)

# Predictions
decision_tree_y_pred = decision_tree_model.predict(X_test)
decision_tree_y_proba = decision_tree_model.predict_proba(X_test)

# Evaluation
print("Decision Tree Accuracy:", accuracy_score(y_test, decision_tree_y_pred))
print("\nDecision Tree Classification Report:")
print(classification_report(y_test, decision_tree_y_pred, target_names=["Dropout", "Enrolled", "Graduate"]))
```

```python
# ============================================================
# CONFUSION MATRIX FUNCTION
# ============================================================

def plot_confusion_matrix(y_true, y_pred, model_name):
    """
    This function plots the confusion matrix for a trained model.
    """
    cm = confusion_matrix(y_true, y_pred)

    disp = ConfusionMatrixDisplay(
        confusion_matrix=cm,
        display_labels=["Dropout", "Enrolled", "Graduate"]
    )

    plt.figure(figsize=(6, 5))
    disp.plot(cmap="Blues", values_format="d")
    plt.title(f"Confusion Matrix - {model_name}")
    plt.show()
```

```python
# ============================================================
# CONFUSION MATRICES FOR ALL MODELS
# ============================================================

plot_confusion_matrix(y_test, ann_y_pred, "Artificial Neural Network")
plot_confusion_matrix(y_test, svm_y_pred, "Support Vector Machine")
plot_confusion_matrix(y_test, decision_tree_y_pred, "Decision Tree")
```

```python
# ============================================================
# MULTI-CLASS ROC-AUC FUNCTION
# ============================================================

def plot_multiclass_roc_auc(y_true, y_proba, model_name):
    """
    This function plots ROC curves and calculates ROC-AUC scores
    for multi-class classification using One-vs-Rest strategy.
    """

    # Define class labels
    classes = [0, 1, 2]
    class_names = ["Dropout", "Enrolled", "Graduate"]

    # Binarize true labels for multi-class ROC calculation
    y_true_binarized = label_binarize(y_true, classes=classes)

    # Calculate overall macro ROC-AUC score
    macro_auc = roc_auc_score(
        y_true_binarized,
        y_proba,
        average="macro",
        multi_class="ovr"
    )

    print(f"{model_name} Macro ROC-AUC Score:", round(macro_auc, 4))

    # Plot ROC curve for each class
    plt.figure(figsize=(8, 6))

    for i, class_name in enumerate(class_names):
        fpr, tpr, _ = roc_curve(y_true_binarized[:, i], y_proba[:, i])
        roc_auc = auc(fpr, tpr)

        plt.plot(
            fpr,
            tpr,
            label=f"{class_name} AUC = {roc_auc:.4f}"
        )

    # Random classifier reference line
    plt.plot([0, 1], [0, 1], linestyle="--", label="Random Classifier")

    plt.title(f"ROC-AUC Curve - {model_name}")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# ============================================================
# ROC-AUC CURVES FOR ALL MODELS
# ============================================================

plot_multiclass_roc_auc(y_test, ann_y_proba, "Artificial Neural Network")
plot_multiclass_roc_auc(y_test, svm_y_proba, "Support Vector Machine")
plot_multiclass_roc_auc(y_test, decision_tree_y_proba, "Decision Tree")
```
